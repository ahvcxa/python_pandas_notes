```python
import pandas as pd
```

```python
data = {
    "Name(Brand)" : ["Audi","BMW","Ford"],
    "Model(year)" : [2024,2012,2008],
    "Cost($)" : [16000,9000,1250]
}
```

```python
df = pd.DataFrame(data, index= ["Car1","Car2","Car3"])
```

```python
print(df)
```

```python
df["FromTheOwner"] = [True,False,True] # Addin new column
```

```python
print(df)
```

```python
newRow = pd.DataFrame([{"Name(Brand)": "Tesla","Model(year)": 2024,"Cost($)": 9000,"FromTheOwner": False}],index= ["Car4"])
df = pd.concat([df,newRow])
print(df)
```
