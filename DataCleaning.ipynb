```python
import pandas as pd 
```

```python
df = pd.read_csv("data.csv")
```

```python
# 1. Drop irrelevant colums 
df = df.drop(columns=["Legendary","No"])
print(df[:10])
```

```python
# 2. Handle missing data 
df = df.dropna(subset= ["Type2"]) # delete not aviable datas 
df = df.fillna({"Type2": "None"}) # fill not avialbe datas 
print(df[:10])
```

```python
# 3. fix inconsistent values
df["Type1"]= df["Type1"].replace({"Grass":"GRASS","Fire":"FIRE"})
print(df[:10])
```

```python
# 4.Standardize text 
df["Name"] =df["Name"].str.lower()
print(df[:10])
```

```python
# 5.Fix data types
df = pd.read_csv("data.csv")
df["legendary"] = df["Legendary"].astype(bool)
print(df[:10])
```
